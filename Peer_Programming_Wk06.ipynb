```python

# dataset

products = ["Sankofa Foods", "Jamestown Coffee", "Bioko Chocolate", "Blue Skies Ice Cream",
            "Fair Afric Chocolate", "Kawa Moka Coffee", "Aphro Spirit", "Mensado Bissap", "Voltic"]

prices = [30, 25, 40, 20, 20, 35, 45, 50, 35]

last_week = [2, 3, 5, 8, 4, 4, 6, 2, 9]

new_price_list = []


# Program Menu

while True:
    print('\n\nWelcome to Azubi Store Back Office')
    operation = input('\nPlease Enter an Operation \n'
                      '1. Consult Total Price Average\n'
                      '2. View Price Review\n'
                      '3. Consult Total Revenue\n'
                      '4. Consult Daily Average Revenue\n'
                      '5. View Products Less Than GHS 30 After Review\n'
                      '6. Quit\n')
    
    if operation == '1':
        # calculate the total price average for all products
        total_price_average = sum(prices) / len(prices)

        print(f'\nTotal Product Price Average: GHS {total_price_average:.2f}\n')
        
    elif operation == '2':
        # create a new price list that reduces the old prices by GHC5

        for new_price in prices:
            new_price_list.append((new_price - 5))
        
        products_new_prices = dict(zip(products, new_price_list))
        print('\nNew Product Prices\n')
        for new_price in products_new_prices:
            print(new_price, ':', products_new_prices[new_price])

    elif operation == '3':
        # calculate the total revenue generated from the products

        total_revenue = 0

        for i in range(len(prices)):
            total_revenue += prices[i] * last_week[i]

        print(f'\nTotal Product Revenue: GHS {total_revenue:.2f}\n')

    elif operation == '4':
        # calculate the average daily revenue generated from the products
        # Assuming Azubi Store works for 6 day a week

        average_daily_revenue = total_revenue / 6

        print(f'\nAverage Daily Revenue: GHS {average_daily_revenue:.2f}')

    elif operation == '5':
        # based on the new prices, which products are less than GHC30
        products_new_prices = dict(zip(products, new_price_list))

        index = 0
        print('\nProducts Price Less Than GHS 30 After Price Review\n')
        for new_product_price in products_new_prices.values():
            if new_product_price < 30:
                print(products[index])
                index += 1

    elif operation == '6':
        break
        
    else:
        print('\nYou Have Entered a Wrong Input, Try Again!')
        continue
```
